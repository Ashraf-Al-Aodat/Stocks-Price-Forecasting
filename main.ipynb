```python
import os
import pandas as pd

dir = 'Stocks'
stock1 = os.listdir(dir + '1')
stock2 = os.listdir(dir + '2')
largest = 'Stocks1/anh.us.txt'


for file in stock2:
    path = dir + '2/' + file
    if(os.stat(largest).st_size < os.stat(path).st_size):
        print(path, os.stat(path).st_size)
        largest= path

for file in stock1:
    path = dir + '1/' + file
    if(os.stat(largest).st_size < os.stat(path).st_size):
        print(path, os.stat(path).st_size)
        largest= path

dir = 'ETFs/ETFs/'
stock1 = os.listdir(dir)

for file in stock1:
    path = dir + file
    if(os.stat(largest).st_size < os.stat(path).st_size):
        print(path, os.stat(path).st_size)
        largest= path

print(largest,  os.stat(largest).st_size)

df =  pd.read_csv(largest, sep="," , encoding = 'ISO-8859-1' , index_col='Date', parse_dates=['Date'])
df = df.dropna()
df
```

```python
companies = pd.read_csv('companies.csv', encoding = 'ISO-8859-1' , index_col='ticker')
pd.DataFrame(companies.loc['GE'])
```

```python
train = df[:'2012'].iloc[:,3:4]
test = df['2013':].iloc[:,3:4]
```

```python
import matplotlib.pyplot as plt
df["Close"][:'2012'].plot(figsize=(16,4),legend=True)
df["Close"]['2013':].plot(figsize=(16,4),legend=True)
plt.legend(['Training set (Before 2012)','Test set (2013 and beyond)'])
plt.title(' General Electric Company stock price')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

sc = MinMaxScaler(feature_range=(0,1))
train_scaled = sc.fit_transform(train)

x_train = []
y_train = []
for i in range(60, len(train)):
    x_train.append(train_scaled[i-60:i,0])
    y_train.append(train_scaled[i,0])
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],1))
x_train.shape
```

```python
import tensorflow.python.keras as k
# The LSTM architecture
regressor = k.Sequential()
# First LSTM layer with Dropout regularisation
regressor.add(k.layers.LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1],1)))
regressor.add(k.layers.Dropout(0.2))
# Second LSTM layer
regressor.add(k.layers.LSTM(units=50, return_sequences=True))
regressor.add(k.layers.Dropout(0.2))
# Third LSTM layer
regressor.add(k.layers.LSTM(units=50, return_sequences=True))
regressor.add(k.layers.Dropout(0.2))
# Fourth LSTM layer
regressor.add(k.layers.LSTM(units=50))
regressor.add(k.layers.Dropout(0.2))
# The output layer
regressor.add(k.layers.Dense(units=1))
```

```python
regressor.summary()
```

```python
# Compiling the model
regressor.compile(optimizer='rmsprop',loss='mean_squared_error')
# Fitting to the training set
history = regressor.fit(x_train,y_train,epochs=12,batch_size=32)
```

```python
# Some functions to help out
def plot_predictions(test, predicted, name):
    import matplotlib.pyplot as plt
    plt.plot(test, color='red',label='Real {} Stock Price'.format(name))
    
    plt.plot(predicted, color='blue',label='Predicted {} Stock Price'.format(name))
    plt.title('{} Stock Price Prediction'.format(name))
    plt.xlabel('Time')
    plt.legend()
    plt.show()

def return_rmse(test, predicted):
    from sklearn.metrics import mean_squared_error
    import math
    rmse = math.sqrt(mean_squared_error(test, predicted))
    print("The root mean squared error is {}.".format(rmse))
```

```python
# Preparing x_test and predicting the prices

sc_unseen = MinMaxScaler(feature_range=(0,1))
unseen_scaled = sc.fit_transform(test)


x_test_unseen = []
for i in range(60, len(unseen_scaled)):
    x_test_unseen.append(unseen_scaled[i-60:i,0])
x_test_unseen = np.array(x_test_unseen)
x_test_unseen = np.reshape(x_test_unseen, (x_test_unseen.shape[0], x_test_unseen.shape[1],1))
predicted = regressor.predict(x_test_unseen)
predicted = sc.inverse_transform(predicted)
```

```python
# Evaluating the model
return_rmse(test[:len(x_test_unseen)], predicted)
```

```python
# Visualizing the results for the model
plot_predictions(test.Close.values, predicted, 'General Electric Company')
```
